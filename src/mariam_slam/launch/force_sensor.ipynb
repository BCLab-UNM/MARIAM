```python
# Curve fit for the force sensor experiment
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# x values and values for 22k resistor
x = np.array([5,5,5,10,10,10,20,20,20,50,50,50,100,100,100,200,200,200,500,500,500])
y22 = np.array([0.26,0.30,1.30,1.55,1.50,0.90,1.49,2.41,1.80,1.30,1.31,2.65,2.13,2.25,3.63,3.83,3.73,3.71,3.38,2.5,1.47])

# log function
def log(x,a,b,c):
    return a * np.log(b * x) + c

# Create the curve fit
popt, pcov = curve_fit(log,x,y22)
x_fit = np.linspace(min(x),max(x),100)
y_fit = log(x_fit,*popt)

# Plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y22, label='Data', color='red')
plt.plot(x_fit, y_fit, label='Fitted Log Curve', color='blue')
plt.title('Logarithmic Fit to Data')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
