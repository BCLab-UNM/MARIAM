```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Data from Xavier's Laptop

```python
bag1_path = '/home/xavier/projects/datasets/bag_1.csv'

data1 = pd.read_csv(bag1_path)['0.002884156']
data1.info()
data1.mean()
```

```python
# Plot the data
plt.figure(figsize=(20, 6))  # Optional: set figure size
data1.plot(kind='line', marker='.', title='Graph of Values')

# Customize the graph
plt.xlabel('Index')  # Label for x-axis
plt.ylabel('Value')  # Label for y-axis
plt.grid(True)       # Add a grid for better readability

# Show the graph
plt.show()
```
